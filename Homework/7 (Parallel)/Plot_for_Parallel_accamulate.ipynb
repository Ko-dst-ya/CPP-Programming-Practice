```python
stack_x = []
stack_y = []
for i in range(len(data)):
    stack_x.append(data[i][0])
    stack_y.append(data[i][1])
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots()
ax.plot(stack_x, stack_y, color="grey", markersize=5, marker='o', linestyle='')
ax.grid()
plt.show()
```
